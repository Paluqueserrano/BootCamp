### Duplicates

```python
movies = pd.DataFrame(movies)
display(movies.drop_duplicates())
```

```python
movies = pd.DataFrame(movies)
movies['budget_USD'] = movies['budget_USD'].astype(int)
movies['gross_USD'] = movies['gross_USD'].astype(int)
movies['release_date'] = pd.to_datetime(movies['release_date'])
```

### most & least frequent MPAA Rating in the dataset in terms of occurrences

```python
mpaa_counts = movies['mpaa_rating'].value_counts() # count the frequency of MPAA Rating
```

```python
mpaa_counts
```

```python
mpaa_counts.nlargest(n=1) # Most frequent
```

### most and least expensive Genre (take an average of all Budget measures grouped by Genre - use groupBy() method)

```python
genre_budget = movies.groupby('genre')['budget_USD'].mean()
```

```python
genre_budget.nlargest(n=1) # the most expensive
```

```python
genre_budget.nsmallest(n=1) #the least expensive
```

### which Genre is favored the most by the people?

```python
genre_rating = movies.groupby('genre')['rating'].mean("rating") # group by genre and calculate the mean of rating
genre_rating.nlargest(n=1) # get the genre with the highest mean rating
```
