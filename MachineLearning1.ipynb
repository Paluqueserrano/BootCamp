```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
red_wine=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=None)
```

```python
white_wine=pd.read_csv('winequality-white.csv', sep=None)
```

```python
white_wine_named=white_wine.rename(columns={'fixed acidity': 'fixed_acidity_g/dm3','volatile acidity': 'volatile_acidity_g/dm3', 'citric acid': 'citric_acid_g/dm3', 'residual sugar': 'residual_sugar', 'chlorides': 'chlorides_g/dm3','free sulfur dioxide':'free_sulfur_dioxide_mg/dm3','total sulfur dioxide':'total_sulfur_dioxide_mg/dm3', 'density':'density_g/cm3','sulphates':'sulphates_g/dm3', 'alcohol': 'alcohol_%'}, inplace=True)
red_wine_named=red_wine.rename(columns={'fixed acidity': 'fixed_acidity_g/dm3','volatile acidity': 'volatile_acidity_g/dm3', 'citric acid': 'citric_acid_g/dm3', 'residual sugar': 'residual_sugar', 'chlorides': 'chlorides_g/dm3','free sulfur dioxide':'free_sulfur_dioxide_mg/dm3','total sulfur dioxide':'total_sulfur_dioxide_mg/dm3', 'density':'density_g/cm3','sulphates':'sulphates_g/dm3', 'alcohol': 'alcohol_%'}, inplace=True)
```

```python
white_wine.drop_duplicates(inplace=True)
red_wine.drop_duplicates(inplace=True)
```

```python
red_wine.insert(0, 'type', 'red')
white_wine.insert(0, 'type', 'white')
```

```python
wines = pd.concat([white_wine, red_wine], axis=0)
print(wines)
```

```python
labels = ['Red wine', 'White wine']
quantity_red = 1359
quantity_white = 3961
my_explode = [0.2, 0]
font1 = {'family': 'serif', 'color': 'blue', 'size': 20}

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie([quantity_red, quantity_white], labels=labels, autopct='%1.1f%%', startangle=90, shadow=True, explode=my_explode, colors=["red", "blue"])
plt.title('Quantity of Red and White Wines', fontdict=font1)
sns.set_theme(rc={'axes.facecolor':'white', 'figure.facecolor': 'white', 'grid.color': 'black' })
```

```python
# Selecciona las columnas relevantes
colnm = wines.columns.to_list()[2:]

# Calcula la matriz de correlaciÃ³n
corr_matrix = wines[colnm].corr()

# Crea una mÃ¡scara triangular superior
mask = np.triu(np.ones_like(corr_matrix))

# Establece el tamaÃ±o de la figura
plt.figure(figsize=(10, 8))

# Crea el mapa de calor
sns.heatmap(corr_matrix, annot=True, cmap="Reds", fmt=".2f", mask=mask)

# Personaliza la apariencia del mapa de calor
plt.title('wines', fontsize=12)
plt.xticks(rotation=90, fontsize=9)
plt.yticks(rotation=0, fontsize=9)

# Muestra el grÃ¡fico
plt.show()
```

# Encoding

```python
wines1=wines.copy()
```

```python
wines1["type"].replace({"white":1, "red":0}, inplace=True)
```

```python
wines1.reset_index(inplace=True)
```

```python
wines1.drop("index", axis=1, inplace=True)
```

```python
wines1.head()
```

```python
wines1.tail()
```

```python
wines1.info()
```

```python
from sklearn.model_selection import train_test_split
X = wines1.drop('type', axis = 1)
y = wines1['type']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=52)
```

```python
X_train.shape 
y_train.shape 
X_test.shape 
y_test.shape 
```

# Logistic Rgresssion

```python
# import LogisticRegression Model
from sklearn.linear_model import LogisticRegression

# LogisticRegression model
lr_model = LogisticRegression(random_state= 42)
```

```python

lr_model.fit(X_train, y_train)
```

```python
preds = lr_model.predict(X_test)
```

```python
from sklearn.metrics import  accuracy_score 

# get the accourcy between predicted data from the model and the orginal data (Test) 
acc = accuracy_score(y_test, preds)

print(acc)
```

# Random Forest

```python
# import Random Forest Classifier model
from  sklearn.ensemble import RandomForestClassifier

# Building model 
RF_model =  RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state=453)
```

```python
RF_model.fit(X_train, y_train)
```

```python
y_pred = RF_model.predict(X_test)

# Evaluate 
acc = accuracy_score(y_test, y_pred)
print(acc)
```
